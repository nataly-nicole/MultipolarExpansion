Se despliega de manera iterativa el potencial eléctrico y las componentes del campo eléctrico hasta el valor $n=10$.

```python
for n in range(11):
    ph_n = sp.Symbol('\phi^{(%d)} ='%n)
    E_nx = sp.Symbol('E^{(%d)}_{x} ='%n)
    E_ny = sp.Symbol('E^{(%d)}_{y} ='%n)
    E_nz = sp.Symbol('E^{(%d)}_{z} ='%n)
    campos_n = read_campos_n(n)
#
    print(display(ph_n, campos_n[0]))
    print(display(E_nx, campos_n[1]))
    print(display(E_ny, campos_n[2]))
    print(display(E_nz, campos_n[3]))
    print(70*'=')
```
